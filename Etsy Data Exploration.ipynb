```python
import json

def jprint(obj):
    # create a formatted string of the Python JSON object
    text = json.dumps(obj, sort_keys=False, indent=4)
    print(text)

jprint(req.json())
```

```python
#print just the results
resultsofquery = req.json()['results']
jprint(resultsofquery)
```

```python
print(type(resultsofquery))
```

```python
#gather all the listing_ids from the query and put into dataframe called listingid
listingid = []

for d in resultsofquery:
    eachlisting = d['listing_id']
    listingid.append(eachlisting)

print(listingid)
```

```python
listingid = []
views = []
title = []
description = []
price = []
currencycode = []
tags = []
url = []
isdigital = []
taxonomypath = []

for d in resultsofquery:
    eachlisting = d['listing_id']
    listingid.append(eachlisting)
    
    eachview = d['views']
    views.append(eachview)
    
    eachtitle = d['title']
    title.append(eachtitle)
    
    eachdescription = d['description']
    description.append(eachdescription)
    
    eachprice = d['price']
    price.append(eachprice)
    
    eachcurrencycode = d['currency_code']
    currencycode.append(eachcurrencycode)
    
    eachtag = d['tags']
    tags.append(eachtag)

    eachurl = d['url']
    url.append(eachurl)
    
    eachdigital = d['is_digital']
    isdigital.append(eachdigital)
    
    eachtaxonomy = d['taxonomy_path']
    taxonomypath.append(eachtaxonomy)
    

print(listingid, url)
```

```python
import pandas as pd
dict = {'listing_id': listingid, 'views': views, 'title': title, 'description': description, 'price': price, 'currency_code': currencycode, 'tags': tags, 'url':url, 'is_digital': isdigital, 'taxonomy_path': taxonomypath}  
    
df = pd.DataFrame(dict) 
    
df  
```
